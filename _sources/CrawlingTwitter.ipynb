## 2. Import Library SNSCRAPE

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

## 3. Crawling data twitter dengan SNSCRAPE

```python
#Insialisai variable list untuk hasil crawling
list_data = []

#Membuat perulangan untuk melakukan crawling menggunakan snscraper dan di masukkan kedalam list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('ridwankamil').get_items()):
    list_data.append([tweet.date, tweet.content, tweet.user. username, tweet.user.location])
    if i>500:
        break
```

* DataFrame hasil crawling twitter

```python
#Membuat dataframe hasil crawlingmel 
df = pd.DataFrame(list_data, columns=['Datetime', 'Text', 'Username', 'Lokasi'])

df
```
